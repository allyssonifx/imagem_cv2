```python
import cv2

metodos = [cv2.THRESH_BINARY,cv2.THRESH_BINARY_INV,cv2.THRESH_TRUNC,cv2.THRESH_TOZERO,cv2.THRESH_TOZERO_INV]
```

```python
img = cv2.imread("xxt.png")
img_cinza = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
i = 0
for metodo in metodos:
    i += 1
    _,img_tratada = cv2.threshold(img_cinza, 0, 255, metodo or cv2.THRESH_OTSU)
    cv2.imwrite(f'imagens/img{i}.png',img_tratada)
```
